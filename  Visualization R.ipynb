***Visualization***

Plotting

```r
png(file = "chart.png")
plot(1:10, main="My Chart", xlab="The x-axis", ylab="The y-axis")
```

```r
x <- mtcars$wt
y <- mtcars$drat
plot(x,y, xlab="weight",ylab="rear axle ratio")
```

```r
a <- c(2, 8)
b <- c(5, 3)
plot(a, b)
```

Line Graphs

```r
plot(1:10,type="l")
```

```r
l1 <- c(0,8,14,42)
l2 <- c(2,4,6,8)
plot(l1,type = "l",col="blue")
lines(l2,type = "l",col="red")
```

```r
x1 <- c(1, 10)
y1 <- c(1, 10)
x2 <- c(1, 10)
y2 <- c(10, 1)
plot(x1, y1, type = "l")
lines(x2, y2, type="l")
```

Bar Charts

```r
barplot(mtcars$hp)
```

```r
barplot(mtcars$hp,names.arg = rownames(mtcars))
```

```r
data <- c(10, 42, 8, 100)
x <- c("B", "Z", "Y", "A")
barplot(data, names.arg = x)
```

```r
barplot(mtcars$hp,horiz= TRUE)
```

```r
barplot(mtcars$hp , col="blue")
```

Pie Chart

```r
x <- c(8,10,42,14)
y <- c("A","B","C","D")
pie(x,label = y)
```

```r
pie(1:8)
```

```r
x <- tapply(mtcars$hp,mtcars$gear,mean)
labels <- names(x)
pie(x,main="Average HP by Gears")
```

```r
x <- tapply(mtcars$wt, mtcars$vs, length)
pie(x)
```
